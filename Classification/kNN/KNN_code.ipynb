# KNN Decision Boundary Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, neighbors
```

```python
neighbors.__doc__
```

## Function To plot KNN Decision Boundary for different Data Sets.

```python
from matplotlib.colors import ListedColormap

def knn_comparision(data, n_neighbors = 15):
    '''
        Its used to find KNN and Plot the Data
    '''
    
    # data[:, : 2] indicates cols 1,2 i.e col 3 is removed  
    x = data[:, :2]
    
    #data[:,2] indicates col 3
    y = data[:, 2]
    
    #grid cell size
    h= 0.02
    
    # mesh color light red and light blue
    cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
    
    # datapoints colors Dark Red and Dark Blue
    cmap_bold = ListedColormap(['#FF0000','#0000FF'])
    
    
    # Core Classifier: K-NN, check the documentation of neighbors.KNeighborsClassifier()
    # at http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
    
    clf = neighbors.KNeighborsClassifier(n_neighbors)
    clf.fit(x,y)
    
    # indicates min and max of 1st col
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    
    #indicates min and max of 2nd col
    y_min ,y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    
    # we create a mesh grid (x_min,y_min) to (x_max,y_max) with 0.02 grid spaces
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    
    
    # we predict the value (either 0 or 1) for each element in the grid
    z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
   
     # xx.ravel() will give a flatten array

    # np.c_ : Translates slice objects to concatenation along the second axis.
    # > np.c_[np.array([1,2,3]), np.array([4,5,6])]
    # > array([[1, 4],
    #          [2, 5],
    #          [3, 6]])   (source: np.c_ documentation)
    
    #convert back the xx to its original shape (we need it to plot the decission boundry)
    z = z.reshape(xx.shape)
    
    
    
     # pcolormesh will plot the (xx,yy) grid with colors according to the values of Z
    # it looks like decision boundry
    plt.figure()
    plt.pcolormesh(xx, yy, z, cmap=cmap_light)
   
    # scatter plot of with given points
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap_bold)
    
    #defining scale on both axises
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
   
    # set the title
    plt.title('K value = '+str(n_neighbors))
    
    
    plt.show()
    
    
```
